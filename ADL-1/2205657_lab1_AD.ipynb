```python
import numpy as np


np.random.seed(42)  
marks = np.random.randint(0, 101, size=30)

print("Dataset (Student Marks):", marks)


mean_marks = np.mean(marks)
median_marks = np.median(marks)
std_deviation = np.std(marks)
variance = np.var(marks)

print("\nStatistical Analysis:")
print(f"Mean: {mean_marks}")
print(f"Median: {median_marks}")
print(f"Standard Deviation: {std_deviation}")
print(f"Variance: {variance}")


max_marks = np.max(marks)
min_marks = np.min(marks)

print(f"\nMaximum Marks: {max_marks}")
print(f"Minimum Marks: {min_marks}")
```

```python
import pandas as pd


try:
    df = pd.read_csv('titanic.csv')  
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print("Error: 'titanic.csv' file not found. Please upload the file and retry.")
    exit()


print("First 5 rows of the dataset:")
print(df.head())


if 'Survived' in df.columns:
    survived = df[df['Survived'] == 1]
    print("\nNumber of passengers who survived:", len(survived))
else:
    print("\nColumn 'Survived' not found in the dataset.")


if 'Age' in df.columns and 'Fare' in df.columns:
    average_age = df['Age'].mean()
    total_fare = df['Fare'].sum()
    print("\nAverage Age of Passengers:", average_age)
    print("Total Fare Collected:", total_fare)
else:
    print("\nRequired columns ('Age' and 'Fare') not found in the dataset.")


if 'Age' in df.columns:
    df['Age'] = df['Age'].fillna(average_age)
    print("\nMissing values in 'Age' column filled with the average age.")
else:
    print("\nColumn 'Age' not found in the dataset.")


print("\nSummary Statistics:")
print(df.describe())
```

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)  # For consistent results
marks1 = np.random.randint(0, 101, size=50)  # Marks for 50 students
marks2 = np.random.randint(0, 101, size=50)  # Another set of marks for comparison

# Step 2: Plot a histogram
plt.figure(figsize=(10, 5))  # Set figure size
plt.hist(marks1, bins=10, color='blue', alpha=0.7, edgecolor='black')
plt.title("Distribution of Marks (Histogram)")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Step 3: Create a scatter plot
plt.figure(figsize=(10, 5))
plt.scatter(marks1, marks2, color='green', alpha=0.7)
plt.title("Scatter Plot of Two Sets of Marks")
plt.xlabel("Marks Set 1")
plt.ylabel("Marks Set 2")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


np.random.seed(42)  # For reproducibility
dates = pd.date_range(start='2024-01-01', end='2024-01-31')  
temperatures = np.random.randint(5, 25, size=len(dates)) 


data = pd.DataFrame({'Date': dates, 'Temperature': temperatures})
data.set_index('Date', inplace=True) 
print("Time-Series Data:")
print(data.head()) 


data['7-day Rolling Avg'] = data['Temperature'].rolling(window=7).mean()


plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Temperature'], label='Daily Temperature', marker='o', color='blue')
plt.plot(data.index, data['7-day Rolling Avg'], label='7-day Rolling Avg', color='red', linewidth=2)
plt.title("Daily Temperatures and 7-day Rolling Average")
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


dates = pd.date_range(start="2024-01-01", end="2024-12-31", freq="D")
temperatures = np.random.normal(loc=25, scale=5, size=len(dates))  # Random temperatures around 25C


df = pd.DataFrame({'Date': dates, 'Temperature': temperatures})


df.set_index('Date', inplace=True)


print("Time-Series Dataset (First 5 rows):")
print(df.head())


df['7-Day Rolling Average'] = df['Temperature'].rolling(window=7).mean()


plt.figure(figsize=(12, 6))


plt.plot(df.index, df['Temperature'], label='Daily Temperature', alpha=0.5, color='blue')


plt.plot(df.index, df['7-Day Rolling Average'], label='7-Day Rolling Average', color='red')


plt.title('Daily Temperature Trends (2024)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Temperature (C)', fontsize=12)
plt.legend()
plt.grid(True)


plt.show()
```
